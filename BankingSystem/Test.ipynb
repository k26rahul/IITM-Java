````python
import os


def create_super_large_string(file_extension):
  current_directory = os.getcwd()
  string = ""
  for filename in os.listdir(current_directory):
    if os.path.isfile(filename) and filename.endswith(file_extension):
      string += f"\n\nFilename: {os.path.abspath(filename)}\n"
      with open(filename, 'r') as file:
        content = file.read()
      string += f"```\n{content}\n```"
  return string


super_large_string_txt = create_super_large_string('.java')
print(super_large_string_txt)
````
