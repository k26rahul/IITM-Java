# Java Reflection API

## Class Methods

- `Class.forName(String className)`
  - Loads and returns the `Class` object associated with the specified class name.

- `Class.getDeclaredMethods()`
  - Returns an array of `Method` objects representing all the declared methods of the class or interface represented by this `Class` object.

- `Class.getDeclaredFields()`
  - Returns an array of `Field` objects representing all the declared fields of the class or interface represented by this `Class` object.

- `Class.getDeclaredConstructors()`
  - Returns an array of `Constructor` objects representing all the declared constructors of the class represented by this `Class` object.

## Method Methods

- `Method.getName()`
  - Returns the name of the method.

- `Method.getReturnType()`
  - Returns a `Class` object representing the return type of the method.

- `Method.getParameterTypes()`
  - Returns an array of `Class` objects representing the parameter types of the method.

- `Method.invoke(Object obj, Object... args)`
  - Invokes the method represented by this `Method` object, on the specified object with the specified parameters.

## Field Methods

- `Field.getName()`
  - Returns the name of the field.

- `Field.getType()`
  - Returns a `Class` object representing the type of the field.

- `Field.get(Object obj)`
  - Returns the value of the field represented by this `Field` object on the specified object.

- `Field.set(Object obj, Object value)`
  - Sets the value of the field represented by this `Field` object on the specified object.

## Constructor Methods

- `Constructor.getName()`
  - Returns the name of the constructor.

- `Constructor.getParameterTypes()`
  - Returns an array of `Class` objects representing the parameter types of the constructor.

- `Constructor.newInstance(Object... initargs)`
  - Uses the constructor represented by this `Constructor` object to create and initialize a new instance of the constructor's declaring class, with the specified initialization parameters.
