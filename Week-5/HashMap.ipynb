# `HashMap` class:

- `put(K key, V value)`: Adds a key-value pair to the `HashMap`.
- `get(Object key)`: Retrieves the value associated with the specified key.
- `containsKey(Object key)`: Checks if the `HashMap` contains a specified key.
- `containsValue(Object value)`: Checks if the `HashMap` contains a specified value.
- `remove(Object key)`: Removes the key-value pair associated with the specified key.
- `keySet()`: Returns a `Set` view of the keys contained in the `HashMap`.
- `values()`: Returns a `Collection` view of the values contained in the `HashMap`.
- `entrySet()`: Returns a `Set` view of the key-value pairs contained in the `HashMap`.
- `size()`: Returns the number of key-value pairs in the `HashMap`.
- `clear()`: Removes all key-value pairs from the `HashMap`.
- `isEmpty()`: Checks if the `HashMap` is empty.
